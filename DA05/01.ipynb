## Pandas Methods

### Create Series

```python
labels = ['a', 'b', 'c']
array = np.array([1, 2, 3])
series = pd.Series(array, labels)
series
```

```python
print(type(series))
```

```python
series = pd.Series(data = array, index = labels)
series
```

```python
series = pd.Series(data = [1, 2, 3, 4], index = ["Iran", "Canada", "Germany", "France"])
series
```

```python
series.shape
```

```python
series.iloc[0]
```

```python
series["Iran"]
```

### Create DataFrame & Modify

```python
data = {
    "name": ["Ali", "Ahmad", "Sina"],
    "age": [20, 26, 21],
    "city": ["Tehran", "Shiraz", "Qom"]
}
rand_df = pd.DataFrame(data)
rand_df
```

```python
data = [
    ["Ali", "Ahmad", "Sina"],
    [20, 26, 21],
    ["Tehran", "Shiraz", "Qom"]
]
rand_df2 = pd.DataFrame(data, columns=["name", "age", "city"])
rand_df2
```

```python
data = [
    {'name': 'Ali', 'age': 22},
    {'name': 'Sara', 'age': 25},
    {'name': 'Reza', 'age': 30}
]

df = pd.DataFrame(data)
print(df)
```

```python
rand_df3 = pd.DataFrame(data=np.random.rand(5, 4),
                  index=[1, 2, 3, 4, 5],
                  columns=["Age", "Salary", "Talent", "Tax"])
rand_df3
```

```python
df = pd.DataFrame(
    data = {
        "age": [14, 24, 26, 56, 46, 55, 59],
        "salary": [100, 200, 300, 400, 500, 600, 700],
    },
    index=["ali", "Ahmad", "Sara", "Reza", "Vahid", "Soheil", "Hossein"]
)
df
```

#### Add a column

```python
rand_df3['New'] = rand_df3["Age"] + rand_df3["Salary"]
rand_df3
```

```python
rand_df3["New 2"] = np.random.randint(5, 10, 5)
rand_df3
```

#### Add a row

```python
rand_df3.loc["NewRow"] = [100, 200, 300, 400, 500, 600]
rand_df3
```
