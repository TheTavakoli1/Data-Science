```python
import pandas as pd
import numpy as np
from numpy import nan as NA

data = pd.DataFrame([[1., 6.5, 3.], [1., NA, NA], 
                     [NA, NA, NA], [NA, 6.5, 3.]])
data.index = ['A','B','C','D']
data
```

```python
data.isna()
```

```python
data.isna().sum()
```

```python
data.dropna()
```
