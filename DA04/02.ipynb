### Reshape

```python
arr = np.random.random(10)
print(arr)
print("----------")

reshape_arr = arr.reshape((5, 2))
print(reshape_arr)
print("----------")

reshape_arr2 = arr.reshape((2, 5))
print(reshape_arr2)
print("----------")

arr2 = np.random.random(10).reshape((5, 2))
print(arr2)
print("----------")
```

### Mathematical and statistical operations

#### Sum, Min, Max, Argmin, Argmax

```python
rand_arr = np.random.randint(1, 20, 10)
print(rand_arr)
print("----------")

print("Max: ", max(rand_arr))
print("Min: ", np.max(rand_arr))
print("Min: ", min(rand_arr))
print("Min: ", np.min(rand_arr))
print("----------")

print(f"Sum: {sum(rand_arr)}")
print(f"Sum: {np.sum(rand_arr)}")
print("----------")

print(f"Argmin: {np.argmin(rand_arr)}")
print("----------")

print(f"Argmax: {np.argmax(rand_arr)}")
```

#### add, subtract, multiply, divide, power

```python
rand_arr = np.random.randint(1, 20, 10)
print("add")
print(np.add(rand_arr, 1000))
print("----------\n")

print("subtract")
print(np.subtract(rand_arr, 1000))
print("----------\n")

print("Multiply")
print(np.multiply(rand_arr, 1000))
print("----------\n")

print("Divide")
print(np.divide(rand_arr, 1000))
print("----------\n")

print("Power")
print(np.power(rand_arr, 2))
print("----------\n")
```

#### std, var, mean

##### Mean

```python
array = np.arange(1, 22).reshape(3, 7)
array
```

```python
print("mean of all elements")
print(np.mean(array))
print("----------\n")

print("mean per row --> (axis=1)")
print(np.mean(array, axis=1))
print("----------\n")

print("mean per column --> (axis=0)")
print(np.mean(array, axis=0))
print("----------\n")
```

##### Variance

```python
print("variance of all elements")
print(np.var(array))
print("----------\n")

print("variance per row (axis=1)")
print(np.var(array, axis=1))
print("----------\n")

print("variance per column (axis=0)")
print(np.var(array, axis=0))
print("----------\n")
```

##### standard deviation

```python
print("standard deviation of all elements")
print(np.std(array))
print("----------\n")

print("standard deviation per row (axis=1)")
print(np.std(array, axis=1))
print("----------\n")

print("standard deviation per column (axis=0)")
print(np.std(array, axis=0))
print("----------\n")
```

### concatinate, vstack, hstack

```python
arr1 = np.random.randint(6, 10, (3, 4))
arr2 = np.random.randint(1, 5, (3, 4))
print(arr1)
print("--------")
print(arr2)
print("----------")

print("Vstack")
arr3 = np.vstack([arr1, arr2])
print(arr3)
print("----------")

print("hstack")
arr4 = np.hstack([arr1, arr2])
print(arr4)
print("----------")

print("concatenate")
print("Axis = 0")
arr5 = np.concatenate((arr1, arr2), axis=0)
print(arr5)

print("----------")
print("Axis = 1")
arr6 = np.concatenate((arr1, arr2), axis=1)
print(arr6)
```
