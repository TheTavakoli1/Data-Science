<h2 style="color:#306998; font-size:23px; font-weight:700; border-bottom:1px solid #EAF2F8; padding-bottom:10px; margin:0 0 18px;">📖 انتخاب ستون‌ها</h2>
<p style="margin:0 0 14px; text-align:justify;">در Part I یاد گرفتیم DataFrame یک <strong>جدول دوبعدی</strong> است: ردیف‌ها مشاهدات (هر خودرو) و <strong>ستون‌ها</strong> متغیرها (رنگ، قدرت، قیمت و …). حالا می‌خواهیم یاد بگیریم چگونه <strong>یک یا چند ستون</strong> را از این جدول جدا کنیم.</p>
<div style="background:#F8F9FB; border-right:3px solid #4B8BBE; border-radius:6px; padding:12px 16px; margin:0 0 26px;">
    <div style="font-weight:700; color:#306998; margin-bottom:6px;">🧠 مدل ذهنی</div>
    <p style="margin:0; text-align:right;">DataFrame را مثل یک فایل اکسل تصور کنید. <strong>انتخاب ستون</strong> یعنی برداشتن یک یا چند ستون <strong>عمودی</strong> از جدول — مثلاً فقط ستون «قیمت» یا ستون‌های «مدل» و «قدرت». در پانداس این کار روی محور <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">axis=1</code> انجام می‌شود (محور افقی = ستون‌ها).</p>
  </div>
<div style="background:#EAF2F8; border-right:3px solid #4B8BBE; border-radius:6px; padding:12px 16px; margin:18px 0; text-align:right; font-weight:600; font-size:18px; color:#306998;">قبل از انتخاب ستون، باید <strong>نام دقیق</strong> آن را بدانیم. ابزار اول برای این کار <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">df.columns</code> است — در بخش بعد.</div>

</div>

```python
import numpy as np
import pandas as pd
```

```python
cars_dictionary = {
    "model" :[ "Peykan", "BMW X6", "Peugeot 206", "Tiba", "Nissan Junior", "Peykan Vanet", "renault 5"],
    "Color" :["Red", "White", "Black", "White", "Blue", "Red", "Silver"],
    "HP" :[48, 335, 105, 87, 93, 80, 48],
    "Weight (KG)": [990, 2320, 1025, 1027, 1540, 960, 850],
    "price" :[33, 2400, 105, 120, 230, 68, 38]
}
df = pd.DataFrame(cars_dictionary)
print(type(df))
df
```
