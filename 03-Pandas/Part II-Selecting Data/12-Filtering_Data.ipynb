<h2 style="color:#306998; font-size:23px; font-weight:700; border-bottom:1px solid #EAF2F8; padding-bottom:10px; margin:0 0 18px;">📖 فیلتر داده — کاربرد عملی</h2>
<p style="margin:0 0 14px; text-align:justify;">در فصل قبل mask بولی و ترکیب شرط‌ها را یاد گرفتیم. این فصل، همان ایده‌ها را در <strong>سناریوهای تحلیلی</strong> جمع می‌کند: یافتن زیرمجموعه‌ای از داده که برای سؤال مشخصی پاسخ می‌دهد.</p>
<div style="background:#EAF2F8; border-right:3px solid #4B8BBE; border-radius:6px; padding:12px 16px; margin:18px 0; text-align:right; font-weight:600; font-size:18px; color:#306998;">فیلتر = انتخاب ردیف بر اساس منطق کسب‌وکار یا تحلیل — نه برچسب و نه موقعیت خام.</div>
<div style="background:#F8F9FB; border-right:3px solid #4B8BBE; border-radius:6px; padding:12px 16px; margin:0 0 26px;">
    <div style="font-weight:700; color:#306998; margin-bottom:6px;">🧠 مدل ذهنی</div>
    <p style="margin:0; text-align:right;">سؤال تحلیلی → شرط(ها) روی ستون → mask → DataFrame فیلترشده → (اختیاری) انتخاب ستون.</p>
  </div>

<h3 style="color:#306998; font-size:18px; margin:28px 0 6px;">🎯 سناریو ۱: فیلتر تک‌شرطی</h3><p style="margin:0 0 12px; font-size:14px; color:#78828C;">سؤال ساده</p>
<p style="margin:0 0 14px; text-align:justify;"><strong>سؤال:</strong> کدام خودروها قدرت (HP) بالای ۱۰۰ دارند؟</p>
<p style="margin:0 0 14px; text-align:justify;">راه‌حل: <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">df[df["HP"] > 100]</code> یا <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">df.loc[df["HP"] > 100]</code></p>
<p style="margin:10px 0 4px; text-align:justify;">خودروهای با وزن کمتر از ۱۱۰۰ کیلوگرم را فیلتر کنید. چند ردیف باقی می‌ماند؟</p>

<h3 style="color:#306998; font-size:18px; margin:28px 0 6px;">🎯 سناریو ۲: فیلتر چندشرطی</h3><p style="margin:0 0 12px; font-size:14px; color:#78828C;">AND / OR</p>
<p style="margin:0 0 14px; text-align:justify;"><strong>سؤال:</strong> خودروهای اقتصادی با HP معقول — مثلاً قیمت زیر ۱۵۰ <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">and</code> HP بالای ۵۰.</p>
<p style="margin:0 0 14px; text-align:justify;">راه‌حل: <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">(df["price"] < 150) & (df["HP"] > 50)</code></p>
<p style="margin:0 0 14px; text-align:justify;"><strong>سؤال:</strong> خودروهای لوکس یا پرقدرت — قیمت بالای ۵۰۰ <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">or</code> HP بالای ۲۰۰.</p>
<p style="margin:0 0 14px; text-align:justify;">راه‌حل: <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">(df["price"] > 500) | (df["HP"] > 200)</code></p>
<p style="margin:10px 0 4px; text-align:justify;">دو فیلتر بالا را روی DataFrame خودروها اجرا کنید و نتایج را با <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">shape[0]</code> مقایسه کنید.</p>

<p style="margin:0 0 14px; text-align:justify;">بعد از فیلتر، معمولاً همه ستون‌ها لازم نیست. مثلاً فقط نام و قیمت خودروهای گران:</p>
<p style="margin:0 0 14px; text-align:justify;"><code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">df.loc[df["price"] > 200, ["model", "price"]]</code> — یک خط، خوانا و بدون زنجیره‌ی <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">SettingWithCopy</code>.</p>
<p style="margin:10px 0 4px; text-align:justify;">خودروهای قرمز را فیلتر کنید و فقط <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">"model"</code> و <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">"Color"</code> را نمایش دهید.</p>
