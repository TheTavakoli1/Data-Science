<h2 style="color:#306998; font-size:23px; font-weight:700; border-bottom:1px solid #EAF2F8; padding-bottom:10px; margin:0 0 18px;">📖 loc — انتخاب با برچسب</h2>

<p style="margin:0 0 14px; text-align:justify;"><code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">loc</code> ابزاری <strong>برچسب‌محور</strong> (label-based) است: برای انتخاب ردیف، ستون، یا هر دو، از نام یا برچسب استفاده می‌کند — نه از موقعیت عددی.</p>

<div style="background:#EAF2F8; border-right:3px solid #4B8BBE; border-radius:6px; padding:12px 16px; margin:18px 0; text-align:right; font-weight:600; font-size:18px; color:#306998;">قانون loc: هر چیزی که داخل <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">loc</code> می‌گذارید باید یک برچسب باشد — برچسب ردیف (index)، نام ستون، یا فهرستی از این برچسب‌ها.</div>

<div style="background:#F8F9FB; border-right:3px solid #4B8BBE; border-radius:6px; padding:12px 16px; margin:0 0 26px;">
  <div style="font-weight:700; color:#306998; margin-bottom:6px;">🧠 مدل ذهنی</div>
  <p style="margin:0; text-align:right;"><code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">df.loc[برچسب_ردیف, برچسب_ستون]</code> — اگر فقط یکی از این دو را بدهید (مثلاً <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">df.loc["BMW X6"]</code>)، همان محور فیلتر می‌شود و محور دیگر به‌طور کامل (همه‌ی ستون‌ها) برگردانده می‌شود.</p>

```python
import pandas as pd
import numpy as np
```

```python
cars_dictionary = {
    "model" :[ "Peykan", "BMW X6", "Peugeot 206", "Tiba", "Nissan Junior", "Peykan Vanet", "renault 5"],
    "Color" :["Red", "White", "Black", "White", "Blue", "Red", "Silver"],
    "HP" :[48, 335, 105, 87, 93, 80, 48],
    "Weight (KG)": [990, 2320, 1025, 1027, 1540, 960, 850],
    "price" :[33, 2400, 105, 120, 230, 68, 38]
}
df = pd.DataFrame(cars_dictionary)
print(type(df))
df
```

```python
df.set_index("model", inplace=True)
```

```python
df
```

```python
df.loc['Peugeot 206']
```
