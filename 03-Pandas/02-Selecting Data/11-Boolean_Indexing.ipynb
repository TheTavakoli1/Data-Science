<h2 style="color:#306998; font-size:23px; font-weight:700; border-bottom:1px solid #EAF2F8; padding-bottom:10px; margin:0 0 18px;">📖 Boolean Indexing — انتخاب با شرط</h2>
<p style="margin:0 0 14px; text-align:justify;">تا اینجا ردیف را با <strong>برچسب</strong> یا <strong>موقعیت</strong> انتخاب کردیم. حالا روش سوم: انتخاب ردیف‌هایی که یک <strong>شرط</strong> را برآورده می‌کنند — مثلاً خودروهایی با HP بالای ۱۰۰.</p>
<div style="background:#EAF2F8; border-right:3px solid #4B8BBE; border-radius:6px; padding:12px 16px; margin:18px 0; text-align:right; font-weight:600; font-size:18px; color:#306998;">ابتدا یک Series از <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">True</code>/<code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">False</code> (mask) می‌سازیم؛ سپس داخل براکت می‌گذاریم: <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">df[df["HP"] > 100]</code>.</div>
<div style="background:#F8F9FB; border-right:3px solid #4B8BBE; border-radius:6px; padding:12px 16px; margin:0 0 26px;">
    <div style="font-weight:700; color:#306998; margin-bottom:6px;">🧠 مدل ذهنی</div>
    <p style="margin:0; text-align:right;">شرط → mask بولی با همان index ردیف‌ها → DataFrame[mask] → فقط ردیف‌های <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">True</code>.</p>
  </div>
<hr style="border:none; border-top:1px solid #E6E9ED; margin:22px 0;">
<h3 style="color:#306998; font-size:18px; margin:28px 0 6px;">✅ ساخت mask بولی</h3><p style="margin:0 0 12px; font-size:14px; color:#78828C;">مقایسه روی ستون</p>
<p style="margin:0 0 14px; text-align:justify;"><code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">df["HP"] > 100</code> → <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">Series</code> از <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">True</code>/<code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">False</code> — برای هر ردیف، آیا شرط برقرار است؟</p>

  </table>
<p style="margin:10px 0 4px; text-align:justify;">mask بولی برای <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">df["HP"] > 80</code> بسازید و آن را چاپ کنید. چند <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">True</code> دارید؟</p>

<h3 style="color:#306998; font-size:18px; margin:28px 0 6px;">🎯 فیلتر با mask</h3><p style="margin:0 0 12px; font-size:14px; color:#78828C;">df[condition]</p>
<p style="margin:0 0 14px; text-align:justify;"><code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">df[df["HP"] > 100]</code> → فقط ردیف‌هایی که HP > 100.</p>
<p style="margin:0 0 14px; text-align:justify;"><code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">df[df["Color"] == "White"]</code> → فقط خودروهای سفید.</p>
<p style="margin:10px 0 4px; text-align:justify;">خودروهایی با قیمت کمتر از ۱۵۰ را فیلتر کنید. سپس خودروهای قرمز با <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">df["Color"] == "Red"</code>.</p>

<h3 style="color:#306998; font-size:18px; margin:28px 0 6px;">🔗 چند شرط: & ، | ، ~</h3><p style="margin:0 0 12px; font-size:14px; color:#78828C;">عملگرهای بولی پانداس</p>
<p style="margin:0 0 14px; text-align:justify;"><strong>و (AND):</strong> <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">(df["HP"] > 80) & (df["price"] < 200)</code> — هر دو شرط باید برقرار باشد.</p>
<p style="margin:0 0 14px; text-align:justify;"><strong>یا (OR):</strong> <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">(df["Color"] == "Red") | (df["Color"] == "Blue")</code>.</p>
<p style="margin:0 0 14px; text-align:justify;"><strong>نه (NOT):</strong> <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">~(df["Color"] == "White")</code> — رنگ سفید نیست.</p>
<div style="background:#FFF8F0; border-right:3px solid #E6A23C; border-radius:6px; padding:12px 16px; margin:18px 0;">
    <div style="font-weight:700; color:#B7791F; margin-bottom:6px;">⚠️ اشتباه رایج</div>
    <p style="margin:0; text-align:right;">از <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">and</code> و <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">or</code> پایتون بین دو Series استفاده نکنید — خطا می‌دهد. حتماً <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">&</code>، <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">|</code>، <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">~</code> و <strong>پرانتز</strong> دور هر شرط: <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">(cond1) & (cond2)</code>.</p>
  </div>
<p style="margin:10px 0 4px; text-align:justify;">خودروهایی با <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">HP > 80</code> <strong>و</strong> قیمت کمتر از ۲۵۰ را پیدا کنید. سپس قرمز <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">یا</code> آبی.</p>

<h3 style="color:#306998; font-size:18px; margin:28px 0 6px;">📋 ترکیب mask و انتخاب ستون</h3><p style="margin:0 0 12px; font-size:14px; color:#78828C;">فیلتر + ستون</p>
<p style="margin:0 0 14px; text-align:justify;">ابتدا ردیف فیلتر، بعد ستون: <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">df[df["HP"] > 100]["model"]</code> — زنجیره‌ای (ترتیب مهم است).</p>
<p style="margin:0 0 14px; text-align:justify;">یا با <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">loc</code>: <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">df.loc[df["HP"] > 100, ["model", "HP", "price"]]</code> — خواناتر برای چند ستون.</p>
<p style="margin:10px 0 4px; text-align:justify;">برای خودروهای با <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">HP > 100</code> فقط ستون‌های <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">"model"</code> و <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">"price"</code> را با <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">loc</code> بگیرید.</p>
