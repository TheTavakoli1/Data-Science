<h2 style="color:#306998; font-size:23px; font-weight:700; border-bottom:1px solid #EAF2F8; padding-bottom:10px; margin:0 0 18px;">📖 انتخاب ردیف‌ها</h2>
<p style="margin:0 0 14px; text-align:justify;">در فصل قبل یاد گرفتیم چگونه <strong>ستون</strong> (خط عمودی جدول) انتخاب کنیم. حالا نوبت <strong>ردیف</strong> است: هر ردیف یک <strong>خط افقی</strong> است که اطلاعات <strong>یک خودرو</strong> را نگه می‌دارد — مدل، رنگ، قدرت، وزن و قیمت با هم.</p>
<div style="background:#F8F9FB; border-right:3px solid #4B8BBE; border-radius:6px; padding:12px 16px; margin:0 0 26px;">
    <div style="font-weight:700; color:#306998; margin-bottom:6px;">🧠 مدل ذهنی</div>
    <p style="margin:0; text-align:right;">در اکسل، <strong>ردیف ۳</strong> یعنی یک مشاهده کامل (یک خودرو). <strong>انتخاب ردیف</strong> یعنی برداشتن آن خط افقی از جدول. در پانداس این کار روی محور <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">axis=0</code> است (محور عمودی = ردیف‌ها). برچسب هر ردیف در <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">df.index</code> ذخیره می‌شود.</p>
  </div>
<div style="background:#EAF2F8; border-right:3px solid #4B8BBE; border-radius:6px; padding:12px 16px; margin:18px 0; text-align:right; font-weight:600; font-size:18px; color:#306998;">ستون سمت <strong>چپ</strong> جدول (اعداد ۰، ۱، ۲، …) برچسب ردیف است — نه یک ستون داده. این برچسب‌ها را با <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">df.index</code> می‌بینیم.</div>
<p style="margin:10px 0 4px; text-align:justify;">همان DataFrame خودروها را بسازید. <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">df.head(2)</code> را اجرا کنید. ردیفی که index آن <strong>۰</strong> است مربوط به کدام خودرو است؟ ردیف index=<strong>1</strong> چطور؟</p>
