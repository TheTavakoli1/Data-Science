  <h2 style="color:#1F4E79; font-size:23px; font-weight:700; border-bottom:2px solid #EAF3FA; padding-bottom:10px; margin:0 0 18px;">
    📖 DataFrame از کجا می‌آید؟
  </h2>

  <p style="margin:0 0 14px; text-align:justify;">
    در فصل قبل با <code dir="ltr" style="unicode-bidi:isolate; background:#EEF2F6; color:#2D6E9E; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:15px;">DataFrame</code> آشنا شدیم و فهمیدیم مهم‌ترین ساختار داده در کتابخانه <code dir="ltr" style="unicode-bidi:isolate; background:#EEF2F6; color:#2D6E9E; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:15px;">Pandas</code> است. اما یک سؤال مهم باقی مانده است:
  </p>

  <div style="background:#EAF3FA; border-right:3px solid #2D6E9E; border-radius:8px; padding:14px 18px; margin:18px 0; text-align:right; font-weight:600; font-size:18px; color:#1F4E79;">
    اگر <code dir="ltr" style="unicode-bidi:isolate; background:#FFFFFF; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:16px; font-weight:400; color:#2D6E9E;">DataFrame</code> یک فایل نیست، پس چگونه ساخته می‌شود؟
  </div>

  <p style="margin:0 0 14px; text-align:justify;">
    پاسخ این سؤال اهمیت زیادی دارد، چون تقریباً هر کاری که بعداً با پانداس انجام می‌دهیم، روی همین ساختار انجام می‌شود.
  </p>

  <p style="margin:0 0 10px; text-align:justify;">
    به‌طور کلی یک <code dir="ltr" style="unicode-bidi:isolate; background:#EEF2F6; color:#2D6E9E; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:15px;">DataFrame</code> از دو منبع ساخته می‌شود:
  </p>

  <ul style="list-style:none; padding:0; margin:0 0 16px;">
    <li style="position:relative; padding-right:20px; margin-bottom:8px;">
      <span style="position:absolute; right:0; color:#2D6E9E;">▸</span>
      داده‌هایی که از قبل داخل برنامه هستند؛ مثل <code dir="ltr" style="unicode-bidi:isolate; background:#EEF2F6; color:#2D6E9E; padding:0 5px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:15px;">Dictionary</code>، <code dir="ltr" style="unicode-bidi:isolate; background:#EEF2F6; color:#2D6E9E; padding:0 5px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:15px;">List</code>، <code dir="ltr" style="unicode-bidi:isolate; background:#EEF2F6; color:#2D6E9E; padding:0 5px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:15px;">NumPy Array</code> یا <code dir="ltr" style="unicode-bidi:isolate; background:#EEF2F6; color:#2D6E9E; padding:0 5px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:15px;">Series</code>
    </li>
    <li style="position:relative; padding-right:20px;">
      <span style="position:absolute; right:0; color:#2D6E9E;">▸</span>
      داده‌هایی که در فایل یا پایگاه داده ذخیره شده‌اند؛ مثل <code dir="ltr" style="unicode-bidi:isolate; background:#EEF2F6; color:#2D6E9E; padding:0 5px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:15px;">CSV</code>، <code dir="ltr" style="unicode-bidi:isolate; background:#EEF2F6; color:#2D6E9E; padding:0 5px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:15px;">Excel</code>، <code dir="ltr" style="unicode-bidi:isolate; background:#EEF2F6; color:#2D6E9E; padding:0 5px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:15px;">JSON</code> یا دیتابیس
    </li>
  </ul>

  <p style="margin:0 0 20px; text-align:justify;">
    در این فصل روی حالت اول تمرکز می‌کنیم: ساخت <code dir="ltr" style="unicode-bidi:isolate; background:#EEF2F6; color:#2D6E9E; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:15px;">DataFrame</code> از ساختارهای موجود در پایتون. ورود مستقیم داده از فایل و پایگاه داده را در فصل‌های بعد می‌بینیم.
  </p>

  <div style="background:#F7F8FA; border:1px solid #E3E6EA; border-right:3px solid #2D6E9E; border-radius:8px; padding:14px 18px; margin:0 0 14px;">
    <div style="font-weight:700; color:#1F4E79; margin-bottom:6px;">🧠 مدل ذهنی</div>
    <p style="margin:0; text-align:right;">
      به <code dir="ltr" style="unicode-bidi:isolate; background:#FFFFFF; color:#2D6E9E; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:15px;">DataFrame</code> مثل یک «جدول استاندارد» نگاه کنید. فرقی نمی‌کند داده اولیه از یک <code dir="ltr" style="unicode-bidi:isolate; background:#FFFFFF; color:#2D6E9E; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:15px;">Dictionary</code> آمده باشد یا یک فایل <code dir="ltr" style="unicode-bidi:isolate; background:#FFFFFF; color:#2D6E9E; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:15px;">CSV</code>؛ پانداس همه را به یک ساختار یکسان تبدیل می‌کند تا با مجموعه‌ای ثابت از توابع روی آن کار کنید.
    </p>
  </div>

  <div style="background:#FFF6E0; border:1px solid #F2E2B3; border-right:3px solid #F0B429; border-radius:8px; padding:14px 18px; margin:0 0 22px;">
    <div style="font-weight:700; color:#8A5A00; margin-bottom:6px;">💡 نکته</div>
    <p style="margin:0; text-align:right;">
      پس از ساخت <code dir="ltr" style="unicode-bidi:isolate; background:#FFFFFF; color:#8A5A00; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:15px;">DataFrame</code>، منبع اولیه‌ی داده دیگر اهمیتی ندارد؛ تقریباً همه‌ی عملیات تحلیل داده از این نقطه به بعد دقیقاً یک‌شکل انجام می‌شود.
    </p>
  </div>

  <h3 style="color:#1F4E79; font-size:18px; margin:0 0 10px;">💻 مثال</h3>
  <p style="margin:0; text-align:justify;">
    یک <code dir="ltr" style="unicode-bidi:isolate; background:#EEF2F6; color:#2D6E9E; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:15px;">DataFrame</code> از یک <code dir="ltr" style="unicode-bidi:isolate; background:#EEF2F6; color:#2D6E9E; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:15px;">Dictionary</code> می‌سازیم و خروجی آن را مشاهده می‌کنیم.

```python
import pandas as pd

cars_dictionary = {
    "model" :[ "Peykan", "BMW X6", "Peugeot 206", "Tiba", "Nissan Junior", "Peykan Vanet", "renault 5"],
    "Color" :["Red", "White", "Black", "White", "Blue", "Red", "Silver"],
    "HP" :[48, 335, 105, 87, 93, 80, 48],
    "Weight (KG)": [990, 2320, 1025, 1027, 1540, 960, 850],
    "price" :[33, 2400, 105, 120, 230, 68, 38]
}

df = pd.DataFrame(cars_dictionary)
print(type(df))
df
```

```python
data = [
    ["Ahmad", 26, "Ph.D", "Tehran"],
    ["Vahid", 29, "Bachelor", "Gilan"],
    ["Soheil", 24, "Diploma", "Tabriz"],
    ["Mehran", 24, "Diploma", "Sarab"],
    ["Bahman", 55, "Master", "Tehran"],
    ["Ariana", 23, "Bachelor", "Tehran"]
    
]
persons_df = pd.DataFrame(data=data, 
                          index=[1, 2, 3, 4, 5, 6], 
                          columns=['Name', 'Age', "Grade", "City"])
print(type(persons_df))
persons_df
```

```python
data = [
    {"Car": "Pride", "Price": 600_000_000, "Model": 1391},
    {"Car": "Dena", "Price": 1_500_000_000, "Model": 1400},
    {"Car": "207", "Price": 1_400_000_000, "Model": 1402},
    {"Car": "206", "Price": 900_000_000, "Model": 1395},
    {"Car": "C200 L", "Price": 12_000_000_000, "Model": 2025}
]
car_df = pd.DataFrame(data=data, 
                      index=range(1, 6, 1))
print(type(car_df))
car_df
```
