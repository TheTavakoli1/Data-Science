<h1 style="color:#1565C0; border-bottom:4px solid #1565C0; padding-bottom:10px;">

<h2>🎯 هدف این فصل</h2>

<p>
در این فصل قصد داریم با کتابخانه <code dir="ltr">Pandas</code> آشنا شویم و بدانیم این کتابخانه چرا ایجاد شده است، چه مشکلاتی را حل می‌کند و مهم‌ترین مفاهیم اولیه آن چیست. همچنین با ساختارهای اصلی مانند <code dir="ltr">DataFrame</code>، <code dir="ltr">Series</code>، <code dir="ltr">Index</code> و <code dir="ltr">Columns</code> آشنا خواهیم شد.
</p>

<hr>

<h2 style="color:#1565C0;">📌 چرا <code dir="ltr">Pandas</code> ساخته شد؟</h2>

<p>
با رشد اینترنت، شبکه‌های اجتماعی، فروشگاه‌های اینترنتی، سامانه‌های بانکی، حسگرها و بسیاری از سیستم‌های اطلاعاتی، حجم بسیار زیادی از داده‌ها تولید شد. علاوه بر افزایش حجم داده‌ها، تنوع آن‌ها نیز بیشتر شد و اطلاعات در قالب فایل‌هایی مانند <code dir="ltr">CSV</code>، <code dir="ltr">Excel</code>، <code dir="ltr">JSON</code>، <code dir="ltr">XML</code> و پایگاه‌های داده ذخیره شدند.
</p>

<p>
ابزارهایی مانند <code dir="ltr">Excel</code> برای تحلیل چنین داده‌هایی کافی نبودند. به همین دلیل کتابخانه <code dir="ltr">Pandas</code> توسعه داده شد تا فرآیند خواندن، پاک‌سازی، تبدیل، تحلیل و آماده‌سازی داده‌ها را بسیار ساده‌تر و سریع‌تر کند.
</p>

<div style="background:#F3E5F5; border-right:7px solid #7B1FA2; padding:18px; border-radius:12px; margin:20px 0;">

<h3>🧠 مدل ذهنی</h3>

<p>
<code dir="ltr">Pandas</code> را می‌توان نسخه برنامه‌نویسی شده <b>Excel</b> در نظر گرفت. با این تفاوت که تمام عملیات به جای کلیک کردن، با استفاده از کدهای پایتون انجام می‌شود و قابلیت‌های بسیار بیشتری برای تحلیل داده دارد.
</p>

</div>

<hr>

<h2 style="color:#1565C0;">📊 <code dir="ltr">DataFrame</code> چیست؟</h2>

<p>
<code dir="ltr">DataFrame</code> مهم‌ترین ساختار داده در <code dir="ltr">Pandas</code> است. این ساختار داده اطلاعات را به صورت یک جدول شامل سطرها (<code dir="ltr">Rows</code>) و ستون‌ها (<code dir="ltr">Columns</code>) نگهداری می‌کند.
</p>

<p>
پانداس می‌تواند داده‌ها را از فایل‌هایی مانند <code dir="ltr">CSV</code>، <code dir="ltr">Excel</code>، <code dir="ltr">JSON</code>، <code dir="ltr">XML</code> و پایگاه‌های داده خوانده و آن‌ها را به یک <code dir="ltr">DataFrame</code> تبدیل کند تا بتوان عملیات تحلیل داده را روی آن انجام داد.
</p>

<div style="background:#FFF8E1; border-right:7px solid #F9A825; padding:18px; border-radius:12px; margin:20px 0;">

<h3>⚠️ اشتباه رایج</h3>

<p>
بسیاری از افراد تصور می‌کنند <code dir="ltr">DataFrame</code> یک فایل است؛ در حالی که <code dir="ltr">DataFrame</code> یک <b>ساختار داده در حافظه</b> است و فایل‌های مختلف تنها به آن تبدیل می‌شوند.
</p>

</div>

<hr>

<h2 style="color:#1565C0;">📈 <code dir="ltr">Series</code> چیست؟</h2>

<p>
در ریاضیات، ساده‌ترین نوع داده یک <b>Scalar</b> (عدد تکی) است. پس از آن بردار یا <b>Vector</b> قرار دارد که مجموعه‌ای از مقادیر در یک بعد است. در <code dir="ltr">Pandas</code> ساختاری به نام <code dir="ltr">Series</code> وجود دارد که شباهت زیادی به یک بردار دارد.
</p>

<p>
هر <code dir="ltr">Series</code> یک آرایه یک‌بعدی است که هر مقدار آن دارای یک <code dir="ltr">Index</code> می‌باشد. همچنین هر ستون از یک <code dir="ltr">DataFrame</code> در واقع یک <code dir="ltr">Series</code> محسوب می‌شود.
</p>

<div style="background:#E8F5E9; border-right:7px solid #2E7D32; padding:18px; border-radius:12px; margin:20px 0;">

<h3>💡 نکته</h3>

<p>
هر <code dir="ltr">DataFrame</code> از کنار هم قرار گرفتن چندین <code dir="ltr">Series</code> تشکیل شده است.
</p>

</div>

<hr>

<h2 style="color:#1565C0;">🔖 <code dir="ltr">Index</code> چیست؟</h2>

<p>
<code dir="ltr">Index</code> مجموعه‌ای از برچسب‌ها است که هر سطر را به صورت یکتا مشخص می‌کند. این برچسب‌ها لزوماً عدد نیستند و می‌توانند نام، کد یا هر مقدار دلخواه دیگری باشند.
</p>

<hr>

<h2 style="color:#1565C0;">📂 <code dir="ltr">Columns</code> چیست؟</h2>

<p>
<code dir="ltr">Columns</code> همان نام ستون‌های یک <code dir="ltr">DataFrame</code> هستند. هر ستون معمولاً یک ویژگی (<code dir="ltr">Feature</code>) یا متغیر (<code dir="ltr">Variable</code>) از داده‌ها را نمایش می‌دهد.
</p>

<h2 style="color:#1565C0;">🕒 چه زمانی از <code dir="ltr">NumPy</code> و چه زمانی از <code dir="ltr">Pandas</code> استفاده کنیم؟</h2>

<ul>

<li>اگر تمرکز شما روی محاسبات عددی، جبر خطی و آرایه‌های چندبعدی است، <code dir="ltr">NumPy</code> انتخاب مناسب‌تری است.</li>

<li>اگر داده‌های شما به صورت جدول هستند و قصد پاک‌سازی، تحلیل، فیلتر کردن یا آماده‌سازی آن‌ها را دارید، <code dir="ltr">Pandas</code> بهترین گزینه است.</li>

</ul>
