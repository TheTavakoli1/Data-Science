# Pandas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame(
    data={
        "Age": [26, 28, 30, 40, 50, 10, 23, 24, 25, 27],
        "Math": [20, 13, 14, 15, 16, 20, 19, 18, 17, 15],
        "Geo": [20, 20, 18, 19, 17, 18.5, 19.5, 20, 16, 15]
    },
    index=["Ahmad", "Farid", "Vahid", "Mohammad", "Abbas", "Shayan", "Soheil", "Mehran", "Farbod", "Hossein"]
)
```

```python
df
```

### Create New column

```python
df["Mean"] = (df["Math"] + df["Geo"]) / 2
```

```python
df
```

```python
df["Salary"] = np.random.randint(1_000, 3_000, 10)
```
