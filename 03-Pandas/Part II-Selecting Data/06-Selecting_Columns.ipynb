<h2 style="color:#306998; font-size:23px; font-weight:700; border-bottom:1px solid #EAF2F8; padding-bottom:10px; margin:0 0 18px;">📖 انتخاب ستون‌ها</h2>
<p style="margin:0 0 14px; text-align:justify;">در Part I یاد گرفتیم DataFrame چیست، چگونه ساخته یا خوانده می‌شود، و با ابزارهای کاوش (مثل <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">columns</code> و <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">head()</code>) ساختار کلی آن را می‌شناسیم. حالا قدم بعدی، <strong>انتخاب بخشی از داده</strong> است — اولین و ساده‌ترین حالت: انتخاب ستون.</p>
<div style="background:#EAF2F8; border-right:3px solid #4B8BBE; border-radius:6px; padding:12px 16px; margin:18px 0; text-align:right; font-weight:600; font-size:18px; color:#306998;">انتخاب ستون = حرکت روی محور axis=1؛ یعنی از بین نام ستون‌ها، یک یا چند ستون را برمی‌داریم.</div>
<div style="background:#F8F9FB; border-right:3px solid #4B8BBE; border-radius:6px; padding:12px 16px; margin:0 0 26px;">
    <div style="font-weight:700; color:#306998; margin-bottom:6px;">🧠 مدل ذهنی</div>
    <p style="margin:0; text-align:right;">یک ستون --> خروجی <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">Series</code> (یک‌بعدی). چند ستون --> خروجی <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">DataFrame</code> کوچک‌تر (دو‌بعدی). این تفاوت در ادامه‌ی همین فصل بارها تکرار می‌شود.</p>
  </div>
<hr style="border:none; border-top:1px solid #E6E9ED; margin:22px 0;">
<h3 style="color:#306998; font-size:18px; margin:28px 0 6px;">📌 انتخاب یک ستون</h3><p style="margin:0 0 12px; font-size:14px; color:#78828C;">سینتکس اصلی: براکت تکی</p>
<p style="margin:0 0 14px; text-align:justify;">رایج‌ترین روش، استفاده از یک براکت و نام ستون به‌صورت رشته است: <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">df["model"]</code>. خروجی یک <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">Series</code> است؛ index آن همان برچسب ردیف‌های DataFrame است و values آن، مقادیر آن ستون.</p>

<p style="margin:10px 0 4px; text-align:justify;">همان DataFrame خودروها را بسازید. ستون <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">"model"</code> را انتخاب کنید و با <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">type()</code> نوع خروجی را بررسی کنید.</p>

```python
import pandas as pd
cars_dictionary = {
    "model" :[ "Peykan", "BMW X6", "Peugeot 206", "Tiba", "Nissan Junior", "Peykan Vanet", "renault 5"],
    "Color" :["Red", "White", "Black", "White", "Blue", "Red", "Silver"],
    "HP" :[48, 335, 105, 87, 93, 80, 48],
    "Weight (KG)": [990, 2320, 1025, 1027, 1540, 960, 850],
    "price" :[33, 2400, 105, 120, 230, 68, 38]
}

df = pd.DataFrame(cars_dictionary)
df
```

```python
print(type(df["model"]))
print('--------')
df["model"]
```
