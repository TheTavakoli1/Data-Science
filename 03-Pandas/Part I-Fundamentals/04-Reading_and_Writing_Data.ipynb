  <p style="margin:0 0 14px; text-align:justify;">
    در دو فصل قبل با <code dir="ltr" style="unicode-bidi:isolate; background:#EEF2F6; color:#2D6E9E; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:15px;">DataFrame</code> و <code dir="ltr" style="unicode-bidi:isolate; background:#EEF2F6; color:#2D6E9E; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:15px;">Series</code> آشنا شدیم و دیدیم چگونه می‌توان آن‌ها را از ساختارهای موجود در برنامه ساخت. اما در دنیای واقعی، داده معمولاً از پیش در جایی ذخیره شده است: یک فایل روی دیسک یا یک پایگاه داده. در این فصل یاد می‌گیریم چگونه داده را از این منابع بخوانیم (Read) و پس از پردازش، دوباره در قالب دلخواه بنویسیم (Write).
  </p>

  <h3 style="color:#1F4E79; font-size:18px; margin:0 0 10px;">📄 CSV</h3>
  <p style="margin:0 0 14px; text-align:justify;">
    CSV (Comma-Separated Values) ساده‌ترین و رایج‌ترین فرمت تبادل داده است: یک فایل متنی که هر خط آن یک ردیف و هر مقدار با کاما (یا جداکننده‌ای دیگر) از مقدار بعدی جدا شده است. سادگی این فرمت باعث پشتیبانی گسترده در تقریباً همه‌ی ابزارها شده، اما هیچ اطلاعاتی درباره‌ی نوع داده یا ساختار سلسله‌مراتبی در خود فایل ذخیره نمی‌شود.
  </p>
  <p style="margin:0 0 4px; text-align:justify;">
    یک فایل CSV را با <code dir="ltr" style="unicode-bidi:isolate; background:#EEF2F6; color:#2D6E9E; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:15px;">read_csv</code> بخوانید و ساختار آن (<code dir="ltr" style="unicode-bidi:isolate; background:#EEF2F6; color:#2D6E9E; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:15px;">.shape</code>، <code dir="ltr" style="unicode-bidi:isolate; background:#EEF2F6; color:#2D6E9E; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:15px;">.head()</code>، <code dir="ltr" style="unicode-bidi:isolate; background:#EEF2F6; color:#2D6E9E; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:15px;">.dtypes</code>) را بررسی کنید.
  </p>

  <h3 style="color:#1F4E79; font-size:18px; margin:26px 0 10px;">🧾 JSON</h3>
  <p style="margin:0 0 14px; text-align:justify;">
    JSON (JavaScript Object Notation) فرمتی سلسله‌مراتبی و متنی است که معمولاً برای تبادل داده بین سرویس‌های وب (API) استفاده می‌شود. برخلاف CSV که ذاتاً دوبعدی (ردیف-ستون) است، JSON می‌تواند ساختارهای تودرتو (nested) مثل لیست داخل دیکشنری را نمایش دهد؛ به همین دلیل گاهی تبدیل آن به یک DataFrame مسطح به مرحله‌ای اضافه نیاز دارد.

    یک DataFrame ساده را با <code dir="ltr" style="unicode-bidi:isolate; background:#EEF2F6; color:#2D6E9E; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:15px;">to_json</code> به رشته‌ی JSON تبدیل کنید و سپس همان رشته را با <code dir="ltr" style="unicode-bidi:isolate; background:#EEF2F6; color:#2D6E9E; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:15px;">read_json</code> دوباره بخوانید.

### Step 1 (Create Simple DataFrame)

```python
cars_dictionary = {
    "model" :[ "Peykan", "BMW X6", "Peugeot 206", "Tiba", "Nissan Junior", "Peykan Vanet", "renault 5"],
    "Color" :["Red", "White", "Black", "White", "Blue", "Red", "Silver"],
    "HP" :[48, 335, 105, 87, 93, 80, 48],
    "Weight (KG)": [990, 2320, 1025, 1027, 1540, 960, 850],
    "price" :[33, 2400, 105, 120, 230, 68, 38]
}

df = pd.DataFrame(cars_dictionary)
```

### Step 2 (Convert to Json)

```python
df.to_json('car.json')
```

### Step 3 (Read Json)

```python
json_df = pd.read_json('car.json')
```

### Step 1 (Create Simple DataFrame)

```python
cars_df = pd.DataFrame(data=[
       ['Peykan', 'Red', 48, 990, 33],
       ['BMW X6', 'White', 335, 2320, 2400],
       ['Peugeot 206', 'Black', 105, 1025, 105],
       ['Tiba', 'White', 87, 1027, 120],
       ['Nissan Junior', 'Blue', 93, 1540, 230],
       ['Nissan Junior', 'Red', 93, 1540, 186],
       ['renault 5', 'Silver', 48, 850, 38]],
            columns=["Model", "Color", "HP", "Weight", "Price"])
cars_df
```

### Step 2 (Convert to XML)

```python
cars_df.to_xml("cars_df.xml", index=False)
```

### Step 3 (Read xml)

```python
xml_df = pd.read_xml('cars_df.xml')
xml_df
```
