```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.DataFrame(data = np.random.randint(0,20,1500).reshape(300,5) , columns = ["fizik" , "hesaban" , "shimi" , "hendese" , "gosaste"])
```

```python
df
```

```python
Q1=np.percentile(df["hesaban"],25)
Q2=np.percentile(df["hesaban"],50)
Q3=np.percentile(df["hesaban"],75)
IQR= Q3 - Q1
lower=Q1 - 1.5*IQR
upper=Q3 + 1.5*IQR
```

```python
(df["hesaban"] > upper).sum()
```

```python
(df["hesaban"] < lower).sum()
```

```python
plt.boxplot(df["hesaban"])
plt.show()
```

```python
plt.scatter(df.index ,df["hesaban"] , label = "hesaban")
plt.plot([1,300],[lower,lower],color="red",linestyle="--")
plt.plot([1,300],[upper,upper],color="green",linestyle="--")
plt.xlabel("student")
plt.ylabel("hesaban")
plt.legend()
plt.show()
```

```python
from sklearn.datasets import load_diabetes
diabetes = load_diabetes()
```

```python
df = pd.DataFrame(diabetes.data , columns = diabetes.feature_names)
df["TARGET"] = diabetes.target
```

```python
df
```

```python
corr = df.corr()
```

```python
corr
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(corr , annot=True, cmap="coolwarm")
plt.title("correlation matrix")
plt.show()
```

```python
plt.scatter(df["bmi"], df["TARGET"])
plt.xlabel("BMI")
plt.ylabel("Target")
plt.show()
```

```python
features = ["age", "sex" , "bp"]
for f in features:
    plt.scatter(df[f], df["TARGET"])
    plt.xlabel(f)
    plt.ylabel(f"scatter {f} vs target")
    plt.show()
```
