# Detecting Outlier

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
seri = pd.Series(np.random.randint(-10,10, size = 30))
seri[10] = -100
seri[27] = 50
np.array(seri)
```

```python
Q1 = seri.quantile(0.25)
Q2 = seri.quantile(0.50)
Q3 = seri.quantile(0.75)
print(f"first quartile = {Q1}")
print(f"third quartile = {Q3}")
```

```python
IQR = Q3 - Q1
upper = Q3 + 1.5 * IQR
lower = Q1 - 1.5 * IQR
```

```python
seri[seri > upper]
```

```python
seri[seri < lower]
```

```python
plt.boxplot(seri)
plt.show
```

```python
IQR = Q3 - Q1
seri.drop(seri[(seri.values < Q1-1.5*IQR) | (seri.values > Q3+1.5*IQR)].index , inplace = True)
np.array(seri)
```

```python
plt.boxplot(seri)
plt.show
```
