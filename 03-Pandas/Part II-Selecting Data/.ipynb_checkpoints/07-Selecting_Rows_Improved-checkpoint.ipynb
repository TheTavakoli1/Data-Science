<h2 style="color:#306998; font-size:23px; font-weight:700; border-bottom:1px solid #EAF2F8; padding-bottom:10px; margin:0 0 18px;">📖 انتخاب ردیف‌ها در Pandas</h2>

<p>
در تحلیل داده، هر ردیف (Row) معمولاً نماینده یک مشاهده، نمونه یا رکورد از دنیای واقعی است.
برای مثال در یک جدول اطلاعات خودروها، هر ردیف می‌تواند مشخصات کامل یک خودرو را شامل نام، سال تولید، قدرت موتور و قیمت نگهداری کند.
</p>

<p>
پس از آشنایی با ستون‌ها، یکی از مهم‌ترین مهارت‌ها در Pandas توانایی انتخاب و بررسی ردیف‌ها است.
تحلیل‌گران داده به‌طور مداوم نیاز دارند یک نمونه خاص را بررسی کنند، چند نمونه را با یکدیگر مقایسه نمایند یا بخشی از داده‌ها را برای تحلیل‌های بعدی استخراج کنند.
</p>

<p>
عملیات انتخاب ردیف‌ها روی محور عمودی جدول انجام می‌شود که در Pandas با <code>axis=0</code> شناخته می‌شود.
درک صحیح مفهوم ردیف و نحوه دسترسی به آن، پایه یادگیری مباحث مهم‌تری مانند فیلتر کردن داده‌ها، گروه‌بندی، ادغام جداول و تحلیل آماری خواهد بود.
</p>

<div style="background:#F4F8FB; border-right:4px solid #306998; padding:12px 16px; border-radius:8px; margin-top:12px;">
هر ردیف = یک مشاهده (Observation) یا نمونه (Sample)<br>
هر ستون = یک ویژگی (Feature) یا متغیر (Variable)
</div>
</div>


```python
import pandas as pd
cars_dictionary = {
    "model" :[ "Peykan", "BMW X6", "Peugeot 206", "Tiba", "Nissan Junior", "Peykan Vanet", "renault 5"],
    "Color" :["Red", "White", "Black", "White", "Blue", "Red", "Silver"],
    "HP" :[48, 335, 105, 87, 93, 80, 48],
    "Weight (KG)": [990, 2320, 1025, 1027, 1540, 960, 850],
    "price" :[33, 2400, 105, 120, 230, 68, 38]
}

df = pd.DataFrame(cars_dictionary)
df
```

```python
df.index
```

```python
df.columns
```

```python
df.set_index("model", inplace=True)
```

```python
df.index
```

<h3 style="color:#306998; font-size:18px; margin:28px 0 6px;">📊 یک ردیف چه شکلی است؟</h3>

<p>
هنگامی که یک ردیف از DataFrame انتخاب می‌شود، Pandas معمولاً آن را به صورت یک شیء از نوع <b>Series</b> نمایش می‌دهد.
این رفتار کاملاً منطقی است؛ زیرا یک ردیف در واقع مجموعه‌ای از مقادیر متعلق به ستون‌های مختلف است.
</p>

<p>
در خروجی Series، نام ستون‌ها نقش Index را بازی می‌کنند و مقادیر ذخیره‌شده در آن ردیف به عنوان داده‌های Series نمایش داده می‌شوند.
به همین دلیل هنگام انتخاب تنها یک ردیف، ساختار جدول به ساختار یک‌بعدی تبدیل می‌شود.
</p>

<p>
اگر به جای یک ردیف، چند ردیف را انتخاب کنیم، ساختار جدولی حفظ شده و خروجی همچنان یک DataFrame خواهد بود.
بنابراین نوع خروجی مستقیماً به تعداد ردیف‌های انتخاب‌شده بستگی دارد.
</p>
</div>



<div dir="rtl" style="font-family:'Vazirmatn','IRANSans',Tahoma,Arial,sans-serif; font-size:17px; line-height:1.9; color:#33404d; max-width:760px; margin:0 auto;">
<h3 style="color:#306998; font-size:18px; margin:28px 0 6px;">⚖️ برچسب (Label) در برابر موقعیت (Position)</h3>

<p>
در Pandas دو روش کاملاً متفاوت برای دسترسی به ردیف‌ها وجود دارد.
درک تفاوت این دو روش یکی از مهم‌ترین مباحث اولیه Pandas است.
</p>

<p>
در روش مبتنی بر <b>Label</b>، Pandas ردیف را بر اساس مقدار Index پیدا می‌کند.
در این حالت مهم نیست ردیف در چه جایگاهی قرار گرفته باشد؛ کافی است برچسب آن را بدانیم.
</p>

<p>
در روش مبتنی بر <b>Position</b>، فقط جایگاه فیزیکی ردیف در جدول اهمیت دارد.
برای مثال عدد 0 به اولین ردیف، عدد 1 به دومین ردیف و به همین ترتیب اشاره می‌کند.
</p>

<p>
بسیاری از خطاهای رایج در Pandas زمانی رخ می‌دهند که Label و Position با یکدیگر اشتباه گرفته شوند.
به همین دلیل قبل از یادگیری <code>loc</code> و <code>iloc</code> باید این تفاوت را به‌خوبی درک کرد.
</p>
</div>



<div dir="rtl" style="font-family:'Vazirmatn','IRANSans',Tahoma,Arial,sans-serif; font-size:17px; line-height:1.9; color:#33404d; max-width:760px; margin:0 auto;">
<h3 style="color:#306998; font-size:18px; margin:28px 0 6px;">✂️ برش ردیف‌ها (Row Slicing)</h3>

<p>
در پروژه‌های واقعی معمولاً تنها یک ردیف انتخاب نمی‌شود.
تحلیل‌گران داده اغلب نیاز دارند بخشی از جدول را استخراج کنند تا روی آن تحلیل، پاک‌سازی یا مدل‌سازی انجام دهند.
</p>

<p>
Pandas امکان انتخاب یک ردیف، چند ردیف مشخص یا یک بازه متوالی از ردیف‌ها را فراهم می‌کند.
به این فرایند Row Slicing یا برش ردیف‌ها گفته می‌شود.
</p>

<p>
برش ردیف‌ها یکی از پرکاربردترین عملیات در تحلیل داده است؛ زیرا تقریباً تمام مراحل بعدی تحلیل، روی زیرمجموعه‌ای از داده‌ها انجام می‌شود نه روی کل جدول.
</p>
</div>

<h3 style="color:#306998; font-size:18px; margin:28px 0 6px;">✅ جمع‌بندی فصل</h3>

<ul>
<li>هر ردیف نماینده یک مشاهده، نمونه یا رکورد در مجموعه داده است.</li>
<li>انتخاب ردیف‌ها روی محور <code>axis=0</code> انجام می‌شود.</li>
<li>هر ردیف دارای یک شناسه به نام Index است.</li>
<li>انتخاب یک ردیف معمولاً خروجی Series تولید می‌کند.</li>
<li>انتخاب چند ردیف معمولاً خروجی DataFrame تولید می‌کند.</li>
<li>دسترسی به ردیف‌ها می‌تواند بر اساس Label یا Position انجام شود.</li>

</ul>

</p>
