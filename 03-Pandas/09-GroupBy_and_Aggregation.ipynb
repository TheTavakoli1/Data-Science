```python
import pandas as pd
```

# 1

```python
students = {
    'name': ['Ahmad', 'Ariana', 'Vahid'],
    'student_id': [1, 2, 3]
}
students_df = pd.DataFrame(data=students)
students_df
```

```python
scores = {
    "score": [18, 20, 17],
    'student_id': [1, 2, 3]
}
scores_df = pd.DataFrame(data=scores)
scores_df
```

```python
pd.merge(students_df, scores_df, on="student_id")
```

# 2

```python
students2 = {
    'name': ['Ahmad', 'Ariana', 'Vahid', 'Peyman'],
    'student_id': [1, 2, 3, 4]
}
students_df2 = pd.DataFrame(data=students2)
students_df2
```

```python
scores2 = {
    "score": [18, 20, 17, 15],
    'id': [1, 2, 3, 4]
}
scores_df2 = pd.DataFrame(data=scores2)
scores_df2
```

```python
pd.merge(students_df2, scores_df2, left_on="student_id", right_on='id')
```

# 3

```python
books = {
    'title': ['HP', 'Hobbit'],
    'authors_id': [0, 1]
}
book_df = pd.DataFrame(books)
book_df
```

```python
authors = {
    'author': ["Rowing", "Tolkien"]
}
authors_df = pd.DataFrame(authors)
authors_df
```

```python
pd.merge(book_df, authors_df, left_on='authors_id', right_index=True)
```
