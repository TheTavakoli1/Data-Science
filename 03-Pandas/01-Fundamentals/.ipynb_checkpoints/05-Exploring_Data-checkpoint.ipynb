<div dir="rtl" style="font-family:'Vazirmatn','IRANSans',Tahoma,Arial,sans-serif; font-size:17px; line-height:1.9; color:#33404d; max-width:760px; margin:0 auto;">
<h2 style="color:#306998; font-size:23px; font-weight:700; border-bottom:1px solid #EAF2F8; padding-bottom:10px; margin:0 0 18px;">
    📖 کاوش در داده‌ها
  </h2>

  <p style="margin:0 0 14px; text-align:justify;">
    در فصل قبل یاد گرفتیم چگونه داده را از منابع مختلف بخوانیم. اما بلافاصله پس از خواندن یک فایل، هنوز نمی‌دانیم دقیقاً با چه داده‌ای طرف هستیم: چند ردیف دارد؟ ستون‌ها چه نامی دارند؟ آیا مقدار خالی وجود دارد؟ این فصل مجموعه‌ای از ابزارهای پانداس را معرفی می‌کند که پیش از هر تحلیلی، یک شناخت اولیه و سریع از داده به ما می‌دهند.
  </p>

  <div style="background:#EAF2F8; border-right:3px solid #4B8BBE; border-radius:6px; padding:12px 16px; margin:18px 0; text-align:right; font-weight:600; font-size:18px; color:#306998;">
    این ابزارها دو نوع‌اند: Attribute (بدون پرانتز؛ فقط اطلاعاتی از پیش موجود را نشان می‌دهد) و Method (با پرانتز؛ عملیاتی روی داده اجرا می‌کند). این تفاوت را در ادامه‌ی همین فصل، برای هر مورد مشخص کرده‌ایم.
  </div>

  <div style="background:#F8F9FB; border-right:3px solid #4B8BBE; border-radius:6px; padding:12px 16px; margin:0 0 26px;">
    <div style="font-weight:700; color:#306998; margin-bottom:6px;">🧠 مدل ذهنی</div>
    <p style="margin:0; text-align:right;">
      این دوازده ابزار در واقع پاسخ به چهار سؤال ساده‌اند: داده چه شکلی است؟ (head، tail، sample) — چقدر بزرگ است؟ (shape، size، ndim) — چه برچسب و نوع داده‌ای دارد؟ (columns، index، dtypes) — و در یک نگاه جامع، وضعیت کلی‌اش چطور است؟ (info، describe، memory_usage)
    </p>
  </div>

  <hr style="border:none; border-top:1px solid #E6E9ED; margin:22px 0;">

  <h3 style="color:#306998; font-size:18px; margin:0 0 6px;">👀 نگاه اول به داده</h3>
  <p style="margin:0 0 12px; font-size:14px; color:#78828C;">Attribute نیستند؛ Method‌اند و باید با پرانتز فراخوانی شوند</p>

  <p style="margin:0 0 14px; text-align:justify;">
    اولین قدم در مواجهه با هر DataFrame جدید، یک نگاه سریع به چند ردیف از آن است، بدون آنکه کل داده (که ممکن است میلیون‌ها ردیف باشد) چاپ شود.
  </p>

        <td style="padding:10px 12px; border:1px solid #E6E9ED;">از ابتدای داده</td>

        <td style="padding:10px 12px; border:1px solid #E6E9ED;"><code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">tail()</code></td>
        <td style="padding:10px 12px; border:1px solid #E6E9ED;">از انتهای داده</td>

        <td style="padding:10px 12px; border:1px solid #E6E9ED;">نمونه‌ی بی‌طرفانه، خصوصاً وقتی داده مرتب‌شده است</td>

  <p style="margin:0 0 4px; text-align:justify;">
    نکته‌ی مهم: اگر داده بر اساس یک ستون مرتب شده باشد (مثلاً تاریخ)، <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:15px;">head()</code> تصویر واقعی و متنوعی از کل داده نمی‌دهد؛ در چنین مواردی <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:15px;">sample()</code> انتخاب بهتری است.
  </p>
  <p style="margin:10px 0 4px; text-align:justify;">
    یک DataFrame دلخواه بسازید و هر سه‌ی <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:15px;">head()</code>، <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:15px;">tail()</code> و <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:15px;">sample()</code> را روی آن اجرا و خروجی‌ها را با هم مقایسه کنید.
  </p>
  <!-- کد را خودت بنویس -->

```python
import pandas as pd
cars_dictionary = {
    "model" :[ "Peykan", "BMW X6", "Peugeot 206", "Tiba", "Nissan Junior", "Peykan Vanet", "renault 5"],
    "Color" :["Red", "White", "Black", "White", "Blue", "Red", "Silver"],
    "HP" :[48, 335, 105, 87, 93, 80, 48],
    "Weight (KG)": [990, 2320, 1025, 1027, 1540, 960, 850],
    "price" :[33, 2400, 105, 120, 230, 68, 38]
}

df = pd.DataFrame(cars_dictionary)
df
```

## head()

```python
df.head()
# df.head(3)
```

## tail()

```python
df.tail()
# df.tail(2)
```

## sample()

```python
df.sample()
```

```python
df.sample(3)
```

```python
df.sample(3, ignore_index=True)
```

```python
df.sample(3, ignore_index=True, random_state=42)
```

  <p style="margin:0 0 4px; text-align:justify;">
    توجه کنید <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:15px;">ndim</code> همان مفهومی است که در فصل Series دیدیم: Series یک‌بعدی و DataFrame دوبعدی است؛ این ویژگی همان تفاوت را به‌صورت عددی گزارش می‌دهد.
  </p>
  <p style="margin:10px 0 4px; text-align:justify;">
    برای همان DataFrame، مقادیر <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:15px;">shape</code>، <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:15px;">size</code> و <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:15px;">ndim</code> را چاپ کنید و بررسی کنید رابطه‌ی size = تعداد ردیف × تعداد ستون برقرار است.
  </p>
  <!-- کد را خودت بنویس -->


```python
data = [
    ["Ahmad", 26, "Ph.D", "Tehran"],
    ["Vahid", 29, "Bachelor", "Gilan"],
    ["Soheil", 24, "Diploma", "Tabriz"],
    ["Mehran", 24, "Diploma", "Sarab"],
    ["Bahman", 55, "Master", "Tehran"],
    ["Ariana", 23, "Bachelor", "Tehran"]
    
]
persons_df = pd.DataFrame(data=data, 
                          index=[1, 2, 3, 4, 5, 6], 
                          columns=['Name', 'Age', "Grade", "City"])
print(type(persons_df))
persons_df
```

## shape

```python
persons_df.shape
```

## size

```python
persons_df.size
```

## ndim

```python
persons_df.ndim
```

    <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:15px;">columns</code> نام ستون‌ها و <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:15px;">index</code> برچسب ردیف‌ها را برمی‌گردانند؛ هر دو از نوع شیء Index هستند، فقط روی محور متفاوتی قرار دارند (columns روی axis=1 و index روی axis=0). <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:15px;">dtypes</code> هم برای هر ستون، نوع داده‌اش را برمی‌گرداند؛ خروجی آن یک Series است که index آن نام ستون‌هاست و مقدارش، dtype متناظر.
