<div dir="rtl" style="font-family:'Vazirmatn','IRANSans',Tahoma,Arial,sans-serif; font-size:17px; line-height:1.9; color:#33404d; max-width:760px; margin:0 auto;">

  <h2 style="color:#1F4E79; font-size:23px; font-weight:700; border-bottom:2px solid #EAF3FA; padding-bottom:10px; margin:0 0 18px;">
    📖 Series چیست؟
  </h2>

  <p style="margin:0 0 14px; text-align:justify;">
    در فصل قبل دیدیم که <code dir="ltr" style="unicode-bidi:isolate; background:#EEF2F6; color:#2D6E9E; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:15px;">DataFrame</code> از داده‌های موجود در برنامه یا از فایل‌ها ساخته می‌شود. اما اگر به ساختار درونی یک <code dir="ltr" style="unicode-bidi:isolate; background:#EEF2F6; color:#2D6E9E; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:15px;">DataFrame</code> دقیق‌تر نگاه کنیم، به یک پرسش بنیادی‌تر می‌رسیم: این جدول از چه واحدهای کوچک‌تری تشکیل شده است؟
  </p>

  <div style="background:#EAF3FA; border-right:3px solid #2D6E9E; border-radius:8px; padding:14px 18px; margin:18px 0; text-align:right; font-weight:600; font-size:18px; color:#1F4E79;">
    <code dir="ltr" style="unicode-bidi:isolate; background:#FFFFFF; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:16px; font-weight:400; color:#2D6E9E;">Series</code> یک آرایه‌ی یک‌بعدی و برچسب‌دار (labeled array) است که می‌تواند داده‌هایی از هر نوع را نگه دارد. هر <code dir="ltr" style="unicode-bidi:isolate; background:#FFFFFF; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:16px; font-weight:400; color:#2D6E9E;">DataFrame</code> در واقع مجموعه‌ای از چند <code dir="ltr" style="unicode-bidi:isolate; background:#FFFFFF; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:16px; font-weight:400; color:#2D6E9E;">Series</code> هم‌ایندکس است.
  </div>

  <p style="margin:0 0 20px; text-align:justify;">
    از منظر جبر خطی، <code dir="ltr" style="unicode-bidi:isolate; background:#EEF2F6; color:#2D6E9E; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:15px;">Series</code> را می‌توان مشابه یک بردار (vector) در نظر گرفت؛ یعنی مجموعه‌ای مرتب از مقادیر در یک بعد. <code dir="ltr" style="unicode-bidi:isolate; background:#EEF2F6; color:#2D6E9E; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:15px;">NumPy</code> همین مفهوم را با یک آرایه‌ی یک‌بعدی پیاده‌سازی می‌کند. اما پانداس یک لایه‌ی مفهومی به آن اضافه می‌کند: به هر عنصر، علاوه بر موقعیتش، یک برچسب معنادار به نام <code dir="ltr" style="unicode-bidi:isolate; background:#EEF2F6; color:#2D6E9E; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:15px;">index</code> نیز نسبت می‌دهد. در نتیجه، در حالی که در یک آرایه‌ی معمولی فقط با موقعیت به عناصر دسترسی داریم، در <code dir="ltr" style="unicode-bidi:isolate; background:#EEF2F6; color:#2D6E9E; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:15px;">Series</code> می‌توان از طریق برچسب‌های معنادار هم به داده دسترسی پیدا کرد.
  </p>

  <h3 style="color:#1F4E79; font-size:18px; margin:0 0 10px;">🧬 اجزای یک Series</h3>

  <p style="margin:0 0 10px; text-align:justify;">هر Series از دو جزء تفکیک‌پذیر تشکیل شده است:</p>

  <ul style="list-style:none; padding:0; margin:0 0 10px;">
    <li style="position:relative; padding-right:20px; margin-bottom:8px;">
      <span style="position:absolute; right:0; color:#2D6E9E;">▸</span>
      <code dir="ltr" style="unicode-bidi:isolate; background:#EEF2F6; color:#2D6E9E; padding:0 5px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:15px;">values</code> — آرایه‌ای از داده‌ها (در سطح پیاده‌سازی، یک NumPy Array)
    </li>
    <li style="position:relative; padding-right:20px;">
      <span style="position:absolute; right:0; color:#2D6E9E;">▸</span>
      <code dir="ltr" style="unicode-bidi:isolate; background:#EEF2F6; color:#2D6E9E; padding:0 5px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:15px;">index</code> — مجموعه‌ای از برچسب‌ها که به هر عنصر values نگاشت می‌شود
    </li>
  </ul>

  <p style="margin:0 0 22px; text-align:justify;">
    این دو جزء همیشه هم‌راستا حرکت می‌کنند: برای هر مقدار در <code dir="ltr" style="unicode-bidi:isolate; background:#EEF2F6; color:#2D6E9E; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:15px;">values</code>، دقیقاً یک برچسب متناظر در <code dir="ltr" style="unicode-bidi:isolate; background:#EEF2F6; color:#2D6E9E; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:15px;">index</code> وجود دارد.
  </p>

  <div style="background:#F7F8FA; border:1px solid #E3E6EA; border-right:3px solid #2D6E9E; border-radius:8px; padding:14px 18px; margin:0 0 22px;">
    <div style="font-weight:700; color:#1F4E79; margin-bottom:6px;">🧠 مدل ذهنی</div>
    <p style="margin:0; text-align:right;">
      به Series مثل یک «دیکشنری مرتب و هم‌نوع» نگاه کنید: <code dir="ltr" style="unicode-bidi:isolate; background:#FFFFFF; color:#2D6E9E; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:15px;">index</code> نقش کلیدها را دارد و <code dir="ltr" style="unicode-bidi:isolate; background:#FFFFFF; color:#2D6E9E; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:15px;">values</code> نقش مقادیر دیکشنری را. تفاوت اصلی این است که Series، برخلاف dict، امکانات محاسباتی و برداری NumPy (مثل جمع، فیلتر و برداری‌سازی عملیات) را نیز به همراه دارد.
    </p>
  </div>

    <div style="font-weight:700; color:#8A5A00; margin-bottom:6px;">💡 نکته</div>
    <p style="margin:0; text-align:right;">
      اگر هنگام ساخت Series پارامتر <code dir="ltr" style="unicode-bidi:isolate; background:#FFFFFF; color:#8A5A00; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:15px;">index</code> مشخص نشود، پانداس به‌طور پیش‌فرض یک <code dir="ltr" style="unicode-bidi:isolate; background:#FFFFFF; color:#8A5A00; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:15px;">RangeIndex</code> (اعداد صحیح از صفر) به آن اختصاص می‌دهد؛ دقیقاً مشابه اندیس‌گذاری پیش‌فرض در List و NumPy Array.
    </p>

  <h3 style="color:#1F4E79; font-size:23px; margin:0 0 10px;">💻 مثال ۱</h3>
  <p style="margin:0; text-align:justify;">
    یک <code dir="ltr" style="unicode-bidi:isolate; background:#EEF2F6; color:#2D6E9E; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:15px;">Series</code> از یک <code dir="ltr" style="unicode-bidi:isolate; background:#EEF2F6; color:#2D6E9E; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:15px;">List</code> می‌سازیم و به‌طور خاص به index پیش‌فرض آن دقت کنید.
  </p>
  <!-- کد را خودت بنویس -->

```python
import pandas as pd
import numpy as np
```

```python
data = ["Iran", "USA", "IRAQ"]
se = pd.Series(data=data)
se
```

<div dir="rtl" style="font-family:'Vazirmatn','IRANSans',Tahoma,Arial,sans-serif; font-size:17px; line-height:1.9; color:#33404d; max-width:760px; margin:0 auto;">

<h3 style="color:#1F4E79; font-size:23px; margin:22px 0 10px;">💻 مثال ۲</h3>
  <p style="margin:0; text-align:justify;">
    اکنون یک <code dir="ltr" style="unicode-bidi:isolate; background:#EEF2F6; color:#2D6E9E; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:15px;">Series</code> از یک <code dir="ltr" style="unicode-bidi:isolate; background:#EEF2F6; color:#2D6E9E; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:15px;">Dictionary</code> بسازید. توجه کنید که کلیدهای دیکشنری به‌طور خودکار به‌عنوان index در نظر گرفته می‌شوند. سپس مقادیر <code dir="ltr" style="unicode-bidi:isolate; background:#EEF2F6; color:#2D6E9E; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:15px;">.values</code>، <code dir="ltr" style="unicode-bidi:isolate; background:#EEF2F6; color:#2D6E9E; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:15px;">.index</code> و <code dir="ltr" style="unicode-bidi:isolate; background:#EEF2F6; color:#2D6E9E; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:15px;">.dtype</code> آن را بررسی کنید.
  </p>
  <!-- کد را خودت بنویس -->

</div>

```python
person_dictionary = {
        "Name": "Ahmad",
        "Family": "Tavakoli",
        "Age": 26,
        "City": "Tehran"
    
}
person_series = pd.Series(person_dictionary)
person_series
```

## .values

```python
person_series.values
```

## .index

```python
person_series.index
```

  <h3 style="color:#1F4E79; font-size:23px; margin:0 0 10px;">💻 مثال 3</h3>
  <p style="margin:0; text-align:justify;">
    یک <code dir="ltr" style="unicode-bidi:isolate; background:#EEF2F6; color:#2D6E9E; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:15px;">Series</code> از دو <code dir="ltr" style="unicode-bidi:isolate; background:#EEF2F6; color:#2D6E9E; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:15px;">List</code> می‌سازیم.
  </p>
  <!-- کد را خودت بنویس -->

```python
index = ['Iran', "USA", 'IRAQ']
population = [80_000_000, 400_000_000, 130_000_000]

pop_se = pd.Series(data=population, index=index)
print(type(pop_se))
pop_se
```
