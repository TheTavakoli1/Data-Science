<h2 style="color:#306998; font-size:23px; font-weight:700; border-bottom:1px solid #EAF2F8; padding-bottom:10px; margin:0 0 18px;">📖 iloc — انتخاب با موقعیت</h2>

<p style="margin:0 0 14px; text-align:justify;"><code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">iloc</code> ابزاری <strong>موقعیت‌محور</strong> (position-based) است: برای انتخاب ردیف، ستون، یا هر دو، از عدد صحیح استفاده می‌کند — نه از برچسب. برچسب index در این حالت اصلاً اهمیتی ندارد.</p>

<div style="background:#EAF2F8; border-right:3px solid #4B8BBE; border-radius:6px; padding:12px 16px; margin:18px 0; text-align:right; font-weight:600; font-size:18px; color:#306998;">قانون iloc: هر چیزی که داخل <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">iloc</code> می‌گذارید باید یک موقعیت عددی باشد — ۰، ۱، ۲، ...، یک slice مثل <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">1:4</code>، یا فهرستی از موقعیت‌ها.</div>

<div style="background:#F8F9FB; border-right:3px solid #4B8BBE; border-radius:6px; padding:12px 16px; margin:0 0 26px;">
  <div style="font-weight:700; color:#306998; margin-bottom:6px;">🧠 مدل ذهنی</div>
  <p style="margin:0; text-align:right;"><code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">df.iloc[موقعیت_ردیف, موقعیت_ستون]</code> — اگر فقط یکی از این دو را بدهید (مثلاً <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">df.iloc[0]</code>)، همان محور فیلتر می‌شود و محور دیگر به‌طور کامل برگردانده می‌شود. مثل اندیس آرایه یا لیست در پایتون، شمارش از صفر شروع می‌شود.</p>
</div>

<h3 style="color:#306998; font-size:18px; margin:28px 0 6px;">🔢 یک ردیف با iloc</h3>
<p style="margin:0 0 12px; font-size:14px; color:#78828C;">با موقعیت عددی (۰، ۱، ۲، ...)</p>

<p style="margin:0 0 14px; text-align:justify;">با <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">df.iloc[0]</code> به ردیفی می‌رسید که در موقعیت ۰ قرار دارد — یعنی اولین ردیف جدول. درست مثل loc، وقتی با یک موقعیت تکی ردیف را انتخاب می‌کنید، خروجی یک <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">Series</code> است که اندیسش همان نام ستون‌های دیتافریم است.</p>

<p style="margin:0 0 14px; text-align:justify;">با <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">df.iloc[2]</code> همیشه ردیفی را می‌گیرید که در موقعیت ۲ قرار دارد (یعنی سومین ردیف) — کاملاً بدون توجه به اینکه برچسب index آن ردیف چیست. این دقیقاً همان تفاوت اصلی loc و iloc است که پیش‌تر دیدید: iloc فقط به موقعیت نگاه می‌کند، نه به برچسب.</p>
