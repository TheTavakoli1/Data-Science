<h2 style="color:#306998; font-size:23px; font-weight:700; border-bottom:1px solid #EAF2F8; padding-bottom:10px; margin:0 0 18px;">📖 انتخاب ردیف‌ها — مقدمه مفهومی</h2>
<p style="margin:0 0 14px; text-align:justify;">پس از انتخاب ستون، نوبت به انتخاب <strong>ردیف</strong> می‌رسد. هر ردیف DataFrame یک مشاهده (observation) کامل است: تمام مقادیر آن ردیف در ستون‌های مختلف.</p>
<div style="background:#EAF2F8; border-right:3px solid #4B8BBE; border-radius:6px; padding:12px 16px; margin:18px 0; text-align:right; font-weight:600; font-size:18px; color:#306998;">انتخاب ردیف = حرکت روی محور axis=0؛ یعنی از بین ردیف‌ها، یک یا چند ردیف را برمی‌داریم.</div>
<div style="background:#F8F9FB; border-right:3px solid #4B8BBE; border-radius:6px; padding:12px 16px; margin:0 0 26px;">
    <div style="font-weight:700; color:#306998; margin-bottom:6px;">🧠 مدل ذهنی</div>
    <p style="margin:0; text-align:right;">دو زبان برای دسترسی به ردیف وجود دارد: <strong>برچسب</strong> (label) از طریق index — و <strong>موقعیت عددی</strong> (position) از صفر تا n-1. در فصل‌های <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">loc</code> و <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">iloc</code> هر کدام را جداگانه یاد می‌گیریم.</p>
  </div>
<hr style="border:none; border-top:1px solid #E6E9ED; margin:22px 0;">
<h3 style="color:#306998; font-size:18px; margin:28px 0 6px;">🏷️ index چیست؟</h3><p style="margin:0 0 12px; font-size:14px; color:#78828C;">برچسب هر ردیف</p>
<p style="margin:0 0 14px; text-align:justify;">هر DataFrame یک <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">index</code> دارد — برچسب ردیف‌ها. پیش‌فرض: اعداد ۰، ۱، ۲، … اما می‌توان index معنادار تعریف کرد، مثلاً نام مدل خودرو: <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">df.set_index("model")</code>.</p>

<p style="margin:10px 0 4px; text-align:justify;">DataFrame خودروها را بسازید. <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">df.index</code> و <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">df.columns</code> را چاپ کنید. سپس با <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">df.set_index("model")</code> index را معنادار کنید و دوباره <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">df.index</code> را ببینید.</p>

<h3 style="color:#306998; font-size:18px; margin:28px 0 6px;">📊 یک ردیف چه شکلی است؟</h3><p style="margin:0 0 12px; font-size:14px; color:#78828C;">خروجی: Series</p>
<p style="margin:0 0 14px; text-align:justify;">وقتی <strong>یک ردیف</strong> انتخاب می‌شود، خروجی معمولاً یک <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">Series</code> است: index آن <strong>نام ستون‌ها</strong> است و هر مقدار، مقدار آن ردیف در ستون مربوط.</p>
<p style="margin:0 0 14px; text-align:justify;">وقتی <strong>چند ردیف</strong> انتخاب می‌شود، خروجی <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">DataFrame</code> است — همان ساختار جدولی با تعداد ردیف کمتر.</p>

  </table>
<p style="margin:10px 0 4px; text-align:justify;">با index پیش‌فرض (۰، ۱، ۲)، ردیف اول را به‌صورت مفهومی در نظر بگیرید: چند ستون دارد؟ index خروجی یک Series از آن ردیف چه چیزهایی است؟ (فعلاً فقط تحلیل ذهنی — در فصل <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">iloc</code> کد می‌زنید.)</p>

<h3 style="color:#306998; font-size:18px; margin:28px 0 6px;">⚖️ برچسب در برابر موقعیت</h3><p style="margin:0 0 12px; font-size:14px; color:#78828C;">پیش‌درآمد loc و iloc</p>
<p style="margin:0 0 14px; text-align:justify;"><strong>دسترسی برچسبی:</strong> می‌دانیم ردیف مورد نظر چه برچسبی دارد — مثلاً <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">"BMW X6"</code> بعد از <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">set_index</code>.</p>
<p style="margin:0 0 14px; text-align:justify;"><strong>دسترسی موقعیتی:</strong> می‌دانیم ردیف چندمین ردیف است — مثلاً «ردیف اول» یعنی موقعیت ۰، بدون توجه به برچسب index.</p>

    <p style="margin:0; text-align:right;">وقتی index پیش‌فرض (۰،۱،۲…) است، <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">df.loc[0]</code> و <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">df.iloc[0]</code> ممکن است یکسان به نظر برسند — اما وقتی index عوض شود، فقط <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">iloc</code> همچنان «ردیف اول» را برمی‌گرداند.</p>

<p style="margin:10px 0 4px; text-align:justify;">بعد از <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">set_index("model")</code>، بنویسید ردیف برچسب <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">"Tiba"</code> در موقعیت عددی چندم قرار دارد؟ (با نگاه به جدول یا <code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">df.index.get_loc("Tiba")</code> اگر بلدید.)</p>

<h3 style="color:#306998; font-size:18px; margin:28px 0 6px;">✂️ برش ردیف (مفهومی)</h3><p style="margin:0 0 12px; font-size:14px; color:#78828C;">یک ردیف، چند ردیف، بازه</p>
<p style="margin:0 0 14px; text-align:justify;">سه الگوی رایج: (۱) یک ردیف — خروجی Series. (۲) لیستی از ردیف‌ها — خروجی DataFrame. (۳) بازه (slice) — چند ردیف پشت‌سرهم.</p>
<p style="margin:0 0 14px; text-align:justify;">تفاوت مهم: در slice <strong>موقعیتی</strong> (<code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">iloc</code>) انتهای بازه شامل نمی‌شود؛ در slice <strong>برچسبی</strong> (<code dir="ltr" style="unicode-bidi:isolate; background:#EFF3F7; color:#4B8BBE; padding:1px 6px; border-radius:4px; font-family:Consolas,'Fira Code',monospace; font-size:14px;">loc</code>) هر دو端 معمولاً شامل می‌شوند — این را در فصل‌های بعد با مثال می‌بینید.</p>
<p style="margin:10px 0 4px; text-align:justify;">برای DataFrame خودروها، لیستی از نام سه خودرو بنویسید که می‌خواهید انتخاب کنید. سپس یک بازه برچسبی (مثلاً از Peykan تا Tiba) و یک بازه موقعیتی (مثلاً ۱ تا ۴) را به‌صورت ذهنی مقایسه کنید: چند ردیف انتخاب می‌شود؟</p>
